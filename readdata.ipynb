```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Set up the Chrome WebDriver
driver = webdriver.Chrome()  # You might need to provide the path if it's not in your PATH
url = "https://www.hongkongairport.com/en/flights/departures/passenger.page"

# Open the webpage
driver.get(url)

# Wait for the content to load
wait = WebDriverWait(driver, 100)  # Wait up to 30 seconds
wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME, "destData")))  # Adjust this class

flight_numbers = []
scheduled_times = []
destinations = []
terminals = []
check_ins = []
transfer_desks = []
gates = []
statuses = []


flight_number = driver.find_elements(By.CLASS_NAME, "flightNo")
scheduled_time = driver.find_elements(By.CLASS_NAME, "timeData") 
destination = driver.find_elements(By.CLASS_NAME, "destData")
terminal = driver.find_elements(By.CLASS_NAME, "terminalData")
check_in = driver.find_elements(By.CLASS_NAME, "checkInData")
transfer_desk = driver.find_elements(By.CLASS_NAME, "transferData")
gate = driver.find_elements(By.CLASS_NAME, "gateData")
status = driver.find_elements(By.CLASS_NAME, "statusData")

destination_counter = 0

for i in range(len(destination)):
    
    # Append data to lists
    if scheduled_time[i] != "":
        flight_numbers.append(flight_number[i].text.strip())
        scheduled_times.append(scheduled_time[i].text.strip())
        destinations.append(destination[destination_counter].text.strip())
        terminals.append(terminal[i].text.strip())
        check_ins.append(check_in[i].text.strip())
        transfer_desks.append(transfer_desk[i].text.strip())
        gates.append(gate[i].text.strip())
        statuses.append(status[i].text.strip())
        destination_counter+=1
    else:
        flight_numbers.append(flight_number[i].text.strip())
        scheduled_times.append("")
        destinations.append("")
        terminals.append("")
        check_ins.append("")
        transfer_desks.append("")
        gates.append("")
        statuses.append("")
# Close the browser
driver.quit()

# Create a DataFrame from the lists
df = pd.DataFrame({
    'Flight Number': flight_numbers,
    'Scheduled Time': scheduled_times,
    'Destination': destinations,
    'Terminal': terminals,
    'Check In': check_ins,
    'Transfer Desk': transfer_desks,
    'Gate': gates,
    'Status': statuses
})

# Display the DataFrame or save it to a CSV file
print(df)
df.to_csv('hong_kong_departures.csv', index=False)
```
